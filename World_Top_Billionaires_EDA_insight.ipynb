####  Fill the null values with the mean, median, or mode of the respective group in the dataset. This approach can be used when there is a logical grouping in the data.

```python
most_frequent_value = df1['Company Name'].mode()[0]

df1['Company Name'].fillna(most_frequent_value, inplace=True)
```

```python
most_frequent_value = df1['Company Relationship'].mode()[0]

df1['Company Relationship'].fillna(most_frequent_value, inplace=True)
```

```python
most_frequent_value = df1['Company Sector'].mode()[0]
df1['Company Sector'].fillna(most_frequent_value, inplace=True)
most_frequent_value = df1['Company Type'].mode()[0]
df1['Company Type'].fillna(most_frequent_value, inplace=True)
most_frequent_value = df1['Demographics Gender'].mode()[0]
df1['Demographics Gender'].fillna(most_frequent_value, inplace=True)
```

```python
most_frequent_value = df1['Wealth Type'].mode()[0]
df1['Wealth Type'].fillna(most_frequent_value, inplace=True)
most_frequent_value = df1['Wealth How Category'].mode()[0]
df1['Wealth How Category'].fillna(most_frequent_value, inplace=True)
most_frequent_value = df1['Wealth How Industry'].mode()[0]
df1['Wealth How Industry'].fillna(most_frequent_value, inplace=True)
```

</div></center>



```python
# Top 5 countries in each year

counts_1996 = df1[df1['Year'] == 1996]['Location Citizenship'].value_counts()
counts_2001 = df1[df1['Year'] == 2001]['Location Citizenship'].value_counts()
counts_2014 = df1[df1['Year'] == 2014]['Location Citizenship'].value_counts()

fig, axs = plt.subplots(1, 3, figsize=(10, 5))

axs[0].pie(counts_1996[:5], labels=counts_1996[:5].index, autopct='%1.1f%%')
axs[0].set_title('Billionaires in 1996')

axs[1].pie(counts_2001[:5], labels=counts_2001[:5].index, autopct='%1.1f%%')
axs[1].set_title('Billionaires in 2001')

axs[2].pie(counts_2014[:5], labels=counts_2014[:5].index, autopct='%1.1f%%')
axs[2].set_title('Billionaires in 2014')

plt.show()
```



  


# <center><div style="font-family: Trebuchet MS; background-color: #0047AB; color: #000000; padding: 12px; line-height: 1;"> Top 10 industries




</div></center>




```python
# Top 10 industries

df = df1.loc[:, ['Wealth Worth In Billions', 'Wealth How Industry']]
df_grouped = df.groupby('Wealth How Industry').sum()
df_sorted = df_grouped.sort_values(by='Wealth Worth In Billions', ascending = False)
df_sorted[:10].plot.bar(color='green')
plt.title('Top 10 Industries with Highest Wealth Worth')
plt.xlabel('Industry')
plt.ylabel('SUM')
plt.show()
```




  


# <center><div style="font-family: Trebuchet MS; background-color: #0047AB; color: #000000; padding: 12px; line-height: 1;">  Mean Wealth by gender




</div></center>




```python
# Mean wealth by gender

mean_wealth = df1.groupby('Demographics Gender')['Wealth Worth In Billions'].mean()
ax = mean_wealth.plot(kind='bar', color=['blue', 'orange', 'green', 'pink'])
ax.set_ylabel('wealth worth in billions')
ax.set_xlabel('gender')
ax.set_title('Mean Wealth by Gender')
plt.show()
```

</div></center>




```python
# Creating group ages

age_labels = ['10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99']
age_bins = [9, 19, 29, 39, 49, 59, 69, 79, 89, 99]
df1['Age_Group'] = pd.cut(df1['Demographics Age'], bins=age_bins, labels=age_labels)
age_groups = df1["Age_Group"].value_counts()
age_groups
```

```python
counts_1996 = df1[df1['Year'] == 1996]['Age_Group'].value_counts()
counts_2001 = df1[df1['Year'] == 2001]['Age_Group'].value_counts()
counts_2014 = df1[df1['Year'] == 2014]['Age_Group'].value_counts()
```

# Top 5 age groups in each year


```python

fig, axs = plt.subplots(1, 3, figsize=(10, 5))

axs[0].pie(counts_1996[:5], labels=counts_1996[:5].index, autopct='%1.1f%%')
axs[0].set_title('Top 5 Age Groups in 1996')

axs[1].pie(counts_2001[:5], labels=counts_2001[:5].index, autopct='%1.1f%%')
axs[1].set_title('Top 5 Age Groups in 2001')

axs[2].pie(counts_2014[:5], labels=counts_2014[:5].index, autopct='%1.1f%%')
axs[2].set_title('Top 5 Age Groups in 2014')

plt.show()
```
